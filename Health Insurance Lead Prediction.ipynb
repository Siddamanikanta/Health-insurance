```python
## remove the unnessasary column
df1=df.drop(['ID','City_Code','Region_Code','Health Indicator','Reco_Policy_Cat'],axis='columns')
df1
```

```python
# average age
df2=df1.copy()
df2["ave_age"]=df2['Upper_Age']|df2['Lower_Age']
df2
```

```python
# consider the only <70 age data
df3=df2[df2.ave_age<70]
df3
```

```python
## drop the columns
df4=df3.drop(['Upper_Age','Lower_Age'],axis='columns')
df4
```

```python
#d=df3['Holding_Policy_Duration'].mode()
#f=df3['Holding_Policy_Type'].mode()
df4['Holding_Policy_Duration'].fillna(1.0,inplace=True)
df4['Holding_Policy_Type'].fillna(3.0,inplace=True)
df4
```

```python
# drop the null values
# check the null values
df5=df4.dropna()
df5.isnull().sum()
```

```python
df5['Response'].replace([0.0,1.0],["no",'yes'],inplace=True)
df5
```

```python
# categorical features
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df5.Accomodation_Type=le.fit_transform(df5.Accomodation_Type)
df5.Reco_Insurance_Type=le.fit_transform(df5.Reco_Insurance_Type)
df5.Is_Spouse=le.fit_transform(df5.Is_Spouse)
df5.Response=le.fit_transform(df5.Response)
df5
```

```python
# scale the data
from sklearn.preprocessing import scale 
mba1=scale(df5.Reco_Policy_Premium)
```

```python
df5['Reco_Policy_Premium']=mba1
df5
```

```python
# build the model
import statsmodels.formula.api as sm
logit_model = sm.logit('Response~Accomodation_Type+Reco_Insurance_Type+Is_Spouse+Holding_Policy_Duration+Holding_Policy_Type+Reco_Policy_Premium+ave_age',data =df5).fit()
logit_model
```

```python
# predict the model
predict=logit_model.predict(pd.DataFrame(df5[['Accomodation_Type','Reco_Insurance_Type','Is_Spouse','Holding_Policy_Duration','Holding_Policy_Type','Reco_Policy_Premium','ave_age']]))
```

```python
# check the accuracy
from sklearn.metrics import accuracy_score 
Score = accuracy_score(df5['Response'],predict > 0.5)
Score
```
